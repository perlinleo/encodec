```python
import os
import soundfile as sf

# Path to the test folder containing WAV files
folder_path = 'test'

# Get all WAV files in the test folder
wav_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]

for wav_file in wav_files:
    file_path = os.path.join(folder_path, wav_file)
    
    # Get file size
    file_size = os.path.getsize(file_path) / (1024 * 1024)  # in MB
    
    # Get bitrate and BPM
    try:
        audio_info = sf.info(file_path)
        bitrate = audio_info.bits / 1000  # in kbps
        bpm = audio_info.samplerate  # BPM might not be available in some WAV files
    except Exception as e:
        print(f"Error reading {wav_file}: {e}")
        bitrate = "N/A"
        bpm = "N/A"
    
    print(f"File: {wav_file}")
    print(f"File Size: {file_size:.2f} MB")
    print(f"Bitrate: {bitrate} kbps")
    print(f"BPM: {bpm}")
    print("-----------------------------")
```
