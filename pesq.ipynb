```python
import os
from peaq import peaq
import soundfile as sf

source_folder = 'test'
mp3_folder = 'mp3'
opus_folder = 'opus'
flac_folder = 'flac'
ecdc_wav_folder = 'ecdc_wav'

fs = 8000  # Set the desired sampling frequency (8000 Hz)

source_files = [file for file in os.listdir(source_folder) if file.endswith('.wav')]
mp3_files = [file for file in os.listdir(mp3_folder) if file.endswith('.mp3')]
opus_files = [file for file in os.listdir(opus_folder) if file.endswith('.opus')]
flac_files = [file for file in os.listdir(flac_folder) if file.endswith('.flac')]
ecdc_wav_files = [file for file in os.listdir(ecdc_wav_folder) if file.endswith('.wav')]

# Function to read audio file and return numpy array and sampling rate
def read_audio(file_path):
    audio_data, sr = sf.read(file_path)
    return audio_data, sr

# PEAQ calculation for MP3 files
for mp3_file in mp3_files:
    source_file = mp3_file.replace('.mp3', '.wav')
    if source_file in source_files:
        print(source_file)
        if source_file == "ar_speak_radio_quran_long_8khz.wav": 
            continue
        source_path = os.path.join(source_folder, source_file)
        mp3_path = os.path.join(mp3_folder, mp3_file)

        ref_audio, _ = read_audio(source_path)
        deg_audio, _ = read_audio(mp3_path)

        score = pesq(fs, ref_audio, deg_audio, 'nb')
        print(f"PEAQ score for {mp3_file}: {score}")

# PEAQ calculation for Opus files (and similar calculations for other formats)
# Adjust folder paths according to your directory structure
```

```python
import os
from pesq import pesq
import soundfile as sf
from tabulate import tabulate

source_folder = 'test'
dest_folders = ['mp3', 'opus', 'flac', 'ecdc_wav']
fs = 8000  # Set the desired sampling frequency (8000 Hz)

source_files = [file for file in os.listdir(source_folder) if file.endswith('.wav')]
dest_files = {folder: [file for file in os.listdir(folder) if file.endswith(('.mp3', '.opus', '.flac', '.wav'))] for folder in dest_folders}

# Function to read audio file and return numpy array and sampling rate
def read_audio(file_path):
    audio_data, sr = sf.read(file_path)
    return audio_data, sr

results = []

# PEAQ calculation for each destination folder
for dest_folder in dest_folders:
    dest_files_list = dest_files[dest_folder]
    for dest_file in dest_files_list:
        source_file = dest_file.replace('.mp3', '.wav').replace('.opus', '.wav').replace('.flac', '.wav')
        if source_file in source_files:
            print(source_file)
            if (source_file == "ar_speak_radio_quran_long_8khz.wav") | (source_file == "song_hip_hop_cut_48khz.wav") | (source_file == "voice_german_pure_cut_48khz.wav"): 
                continue
            source_path = os.path.join(source_folder, source_file)
            dest_path = os.path.join(dest_folder, dest_file)

            ref_audio, _ = read_audio(source_path)
            deg_audio, _ = read_audio(dest_path)

            score = pesq(fs, ref_audio, deg_audio, 'nb')
            results.append([dest_file, score])

# Creating a table
headers = ["File", "PEAQ Score"]
table = tabulate(results, headers=headers, tablefmt="tsv")

print(table)
```
